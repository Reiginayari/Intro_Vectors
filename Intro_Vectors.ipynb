1. Let $\textbf{u} = (1, -2, 4)$ and $\textbf{v} = (3,5,1)$ $\textbf{w} = (2,1,-3)$ find:<br>
a. $3\textbf{u}âˆ’2\textbf{v}=(3,âˆ’6,12)âˆ’(6,10,2)=(âˆ’3,âˆ’16,10)$<br>
b. $5\textbf{u}+3\textbf{v}âˆ’4\textbf{w}=(5,âˆ’10,20)+(9,15,3)âˆ’(8,4,âˆ’12)=(6,1,29)$<br>
c. $\textbf{u}â‹…\textbf{v}=1â‹…3+(âˆ’2)â‹…5+4â‹…1=âˆ’3$ <br>
$\textbf{v}â‹…\textbf{w}=3â‹…2+5â‹…1+1â‹…(âˆ’3)=8$<br>
$\textbf{u}â‹…\textbf{w}=1â‹…2+(âˆ’2)â‹…1+4â‹…(âˆ’3)=âˆ’12$<br>
d. $||\textbf{u}||= \sqrt{1^2+(-2)^2+4^2} = \sqrt{21}$<br>
$||\textbf{v}||=\sqrt{3^2+5^2+1^2}=\sqrt{35}$<br>
$||\textbf{w}||=\sqrt{2^2+1^2+(-3)^2}=\sqrt{14}$<br>
e. $cos Î¸=\frac{u.v}{||u||||v||}=\frac{-3}{\sqrt{21}\sqrt{35}}$<br>
f. $cos Î±=\frac{v.w}{||v||||w||}=\frac{8}{\sqrt{35}\sqrt{14}}$<br>
g. $cos Î²=\frac{u.w}{||u||||w||}=\frac{-12}{\sqrt{21}\sqrt{14}}$<br>
h. $d(\textbf{u,v})=âˆ¥uâˆ’vâˆ¥= \sqrt{(-2)^2+(-7)^2+(3)^2}=\sqrt{62}$<br>
i. $d(\textbf{w,v})=||w-v||=\sqrt{(2-3)^2+(1-5)^2+(-3-1)^2}=\sqrt{34}$<br>
j. $d(\textbf{u,w})=||u-w||=\sqrt{(1-2)^2+(-2-1)^2+(4+3)^2}=\sqrt{84}$<br>
k. $proj(\textbf{u,v})=\frac{u.v}{||v||^2}\textbf{v}$<br>
l. $proj(\textbf{w,v})=\frac{w.v}{||v||^2}\textbf{v}$<br>
m. $proj(\textbf{u,w})=\frac{u.w}{||w||^2}\textbf{w}$<br>


```python
import numpy as np

# Define the vectors
u = np.array([1, -2, 4])
v = np.array([3, 5, 1])
w = np.array([2, 1, -3])

# a. 3u - 2v
result_a = 3 * u - 2 * v

# b. 5u + 3v - 4w
result_b = 5 * u + 3 * v - 4 * w

# c. Dot products
u_dot_v = np.dot(u, v)
v_dot_w = np.dot(v, w)
u_dot_w = np.dot(u, w)

# d. Norms
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)
norm_w = np.linalg.norm(w)

# e. Cosine of angles
cos_theta = u_dot_v / (norm_u * norm_v)
cos_alpha = v_dot_w / (norm_v * np.linalg.norm(w))
cos_beta = u_dot_w / (norm_u * np.linalg.norm(w))

# h. Distance between vectors
d_uv = np.linalg.norm(u - v)
d_wv = np.linalg.norm(w - v)
d_uw = np.linalg.norm(u - w)

# k, l, m. Vector projections
proj_uv = (u_dot_v / (norm_v ** 2)) * v
proj_wv = (v_dot_w / (norm_v ** 2)) * v
proj_uw = (u_dot_w / (norm_w ** 2)) * w

# Print results
print("a. 3u - 2v =", result_a)
print("b. 5u + 3v - 4w =", result_b)
print("c. u dot v =", u_dot_v)
print("   v dot w =", v_dot_w)
print("   u dot w =", u_dot_w)
print("d. ||u|| =", norm_u)
print("   ||v|| =", norm_v)
print("   ||w|| =", norm_w)
print("e. cos(theta) =", cos_theta)
print("f. cos(alpha) =", cos_alpha)
print("g. cos(beta) =", cos_beta)
print("h. d(u, v) =", d_uv)
print("   d(w, v) =", d_wv)
print("   d(u, w) =", d_uw)
print("k. proj(u, v) =", proj_uv)
print("l. proj(w, v) =", proj_wv)
print("m. proj(u, w) =", proj_uw)
```
