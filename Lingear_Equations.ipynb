4. Provide reasons why the given matrices are either a row echelon, a reduced row echelon, or niether both.<br>
a. <br>
\begin{pmatrix}
   1 & 0 & 1 \\
   0 & 0 & 3 \\
   0 & 1 & 0 \\
\end{pmatrix}<br><br>
Because the second row defies the requirements of row-echelon form, this matrix is not in row-echelon form. Each nonzero row's leading entry appears to the right of the preceding row's leading entry in row-echelon form. In this instance, the first row's leading entry is to the left of the second row's leading entry. <br>
b. <br>
\begin{pmatrix}
   7 & 0 & 1 & 0 \\
   0 & 1 & -1 & 4 \\
   0 & 0 & 0 & 0 \\
\end{pmatrix}<br><br>
The form of this matrix is row-echelon. It meets the requirements for row-echelon form.
Any rows with zeros are at the bottom.
Every nonzero row's leading entry appears to the right of the row's preceding leading entry.
In any row that is not zero, the first entry is 1. <br>
c. <br>
\begin{pmatrix}
   0 & 1 & 3 & 0 \\
   0 & 0 & 0 & 1 \\
\end{pmatrix}<br><br>
There is no row-echelon form for this matrix. The row-echelon form's qualities are broken by the second row. If a row has any nonzero items in row-echelon form, the leading entry has to be 1. Although it is not in the first column, 1 is the leading entry in this instance of the second row.<br>
d. <br>
\begin{pmatrix}
   0 & 0 & 0 \\
   0 & 0 & 0 \\
   0 & 0 & 0 \\
\end{pmatrix}<br><br>
There is no row-echelon form for this matrix. It violates the requirement that any zero rows should be at the bottom because it is entirely made up of zero rows. <br>
e. <br>
\begin{pmatrix}

   0 & 0 & 0 & 0 & 0 \\
   0 & 1 & 5 & 0 & 1 \\
\end{pmatrix}<br><br>
The form of this matrix is row-echelon. It meets the requirements for row-echelon form.<br>
f. <br>
\begin{pmatrix}
   0 & 0 & 1 \\
   0 & 1 & 0 \\
   1 & 0 & 0 \\
\end{pmatrix}<br><br>
In reduced row-echelon form (RREF), this matrix is available. The requirements of RREF are met by it: any zero rows are at the bottom. Every nonzero row has a leading entry that is the only nonzero entry in its column. <br>
